1.4 visualizations

```python
#pivot tables of  adress,price and  rooms
pivot=df1.pivot_table(
    values=['Rooms','Price'],
    index='Address',
    aggfunc=['mean', 'sum'])
print(pivot)
```

```python
# histogram for price,rooms,bedroom2,bathroom
df[['Price','Rooms','Bedroom2','Bathroom','Distance','Car']].hist(bins=10, figsize=(12, 6), edgecolor='black')
plt.grid(True)
plt.show()
```

```python
#pie chart of prices per number of bedrooms
Price_totals = df.groupby('Bedroom2')['Price'].sum()
Price_totals.plot(kind='bar')
plt.title('Price per bedrooms')
plt.xlabel('Bedroom2')
plt.ylabel('Price')
plt.show()
```

```python
#pie chart showing number of rooms
category_totals = df.groupby('Rooms')['Price'].sum()
plt.pie(category_totals, labels=category_totals.index, autopct='%1.1f%%', startangle=140)
plt.title('Category Breakdown')
plt.show()
```

```python
#the date sold vs the year  built
df['Date'] = pd.to_datetime(df['Date'])
daily_totals = df.groupby('Date')['YearBuilt'].sum()
plt.plot(daily_totals.index, daily_totals)
plt.title('Date sold vs Yearbuilt')
plt.xlabel('Date')
plt.ylabel('YearBuilt')
plt.xticks(rotation=45)
plt.show()
```

```python
#
heatmap_data = df.pivot_table(values='Bedrooms2', index='Rooms', columns='Price', aggfunc='sum')
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu')
plt.title('Heatmap of Amount by Date and Category')
plt.show()
```
